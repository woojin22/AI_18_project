# **ü•á ÎèÑÏ†Ñ Í≥ºÏ†ú**


Ïó¨Îü¨ Îç∞Ïù¥ÌÑ∞ÏÖãÏùÑ Ìï©Ïπ† ÎïåÏóêÎäî `concat`, `merge`, `join` Îì±Ïùò Ìï®ÏàòÎ•º ÌôúÏö©Ìï† Ïàò ÏûàÏäµÎãàÎã§.

Ïù¥Î≤à Í≥ºÏ†úÎ•º ÌÜµÌï¥ÏÑúÎäî Toy DataÎ•º ÌôúÏö©ÌïòÏó¨ `concat`, `merge`, `join`Ïùò Í≤∞Í≥ºÎ•º ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú ÏÇ¥Ìé¥Î≥¥Í≥†, Í∞Å Ìï®ÏàòÏùò Ï∞®Ïù¥Ï†êÏùÑ ÌååÏïÖÌï¥Î¥ÖÎãàÎã§. ÌäπÌûà ÌååÎùºÎØ∏ÌÑ∞Î•º Î≥ÄÍ≤ΩÌï¥Í∞ÄÎ©∞ ÏûëÎèôÏõêÎ¶¨Î•º ÏûêÏÑ∏Ìûà ÏÇ¥Ìé¥Î≥¥ÏÑ∏Ïöî.



```python
df1 = pd.DataFrame({'a':['a0','a1','a2','a3'],
                   'b':['b0','b1','b2','b3'],
                   'c':['c0','c1','c2','c3']},
                  index = [0,1,2,3])

df2 = pd.DataFrame({'a':['a2','a3','a4','a5'],
                   'b':['b2','b3','b4','b5'],
                   'c':['c2','c3','c4','c5'],
                   'd':['d2','d3','d4','d5']},
                   index = [2,3,4,5])
```

### **Part 2. concat**
- axis ÌååÎùºÎØ∏ÌÑ∞Î•º Ï°∞Ï†ïÌïòÏó¨ Ïó¥Î∞©Ìñ•ÏúºÎ°ú Îëê Îç∞Ïù¥ÌÑ∞ÏÖãÏùÑ Ìï©Ï≥êÎ≥¥ÏÑ∏Ïöî.
- axis ÌååÎùºÎØ∏ÌÑ∞Î•º Ï°∞Ï†ïÌïòÏó¨ ÌñâÎ∞©Ìñ•ÏúºÎ°ú Îëê Îç∞Ïù¥ÌÑ∞ÏÖãÏùÑ Ìï©Ï≥êÎ≥¥ÏÑ∏Ïöî.

```python
result1 = pd.concat([df1,df2], axis=1)
result2 = pd.concat([df1,df2])
result1
```

## **Part 3. merge**
- `a` columnÏùÑ Í∏∞Ï§ÄÏúºÎ°ú how ÌååÎùºÎØ∏ÌÑ∞Ïóê Í∞ÅÍ∞Å 'inner', 'outer', 'left', 'right'Î•º Ï†ÅÏö©ÌïòÏó¨ Îëê Îç∞Ïù¥ÌÑ∞ÏÖãÏùÑ Ìï©Ïπú ÌõÑ, Í∑∏ Ï∞®Ïù¥Î•º ÌôïÏù∏Ìï¥Î≥¥ÏÑ∏Ïöî.

```python
result3 = pd.merge(df1, df2, how='left', on='a')
result4 = pd.merge(df1, df2, how='right', on='a')
result5 = pd.merge(df1, df2, how='outer', on='a')
result6 = pd.merge(df1, df2, how='inner', on='a')


result3
```
