```python
error = movie[movie.Released_Year.str.contains(r'[^0-9]')]
error
```

```python
# findall()ÏùÑ ÌôúÏö©ÌïòÏó¨ Ï∂úÎ†•Ìï¥Î≥º ÏàòÎèÑ ÏûàÏäµÎãàÎã§. 

import re

not_num = []
for year in movie['Released_Year']:
  string = re.compile(r'[\D+]')
  result = string.findall(year)
  if len(result) >= 1:
    not_num.append(year)

not_num
```

```python
# re.searchÏôÄ applyÎ•º ÌôúÏö©ÌïòÏó¨ Îç∞Ïù¥ÌÑ∞Î•º Ï∂îÏ∂úÌï† ÏàòÎèÑ ÏûàÏäµÎãàÎã§. 

import re

def has_errors(inputString):
  return bool(re.search('[\D]', inputString))
```

```python
error_condition = movie.Released_Year.apply(has_errors)
error_names = movie[error_condition]
error_names
```
