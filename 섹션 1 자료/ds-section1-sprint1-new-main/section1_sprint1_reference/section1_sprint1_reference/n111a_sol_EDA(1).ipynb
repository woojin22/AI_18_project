```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 파일 url
url = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/stocks/Travel.xlsx'

# 데이터프레임으로 불러오기
df1 = pd.read_excel(url, sheet_name= 0)
df2 = pd.read_excel(url, sheet_name= 1)
```

```python
df1.head(3)
```

```python
df2.head(3)
```

```python
# 한 줄씩 실행해 보며 데이터프레임이 어떻게 바뀌어 가는지 살펴보세요. 
# column에 특성이 나타나도록 행과 열을 바꿉니다. 
df1 = df1.T
df2 = df2.T
```

```python
# 0번째 row에 있는 컬럼명을 df_columns 변수에 저장해 둡니다. 
df_columns = df1.iloc[0]

# 헤더에 index가 있으므로 이를 컬럼명으로 바꿔줍니다. 
df1.columns = df_columns
df2.columns = df_columns
```

```python
# 헤더에 컬럼명이 잘 들어왔으므로 0번째 row는 더이상 필요하지 않습니다. 
# 슬라이싱을 활용하여 0번째 row는 제외하고 나머지 row만 추출합니다. 
df1 = df1[1:]
df2 = df2[1:]
```

## **Part 2. Missing Value**
**자산총계의 결측치를 처리하세요.** 

앞선 문제에서는 결측치를 단순히 0으로 처리하였으나, 결측치를 처리하는 방법은 그 외에도 여러가지가 있습니다. <br>
다양한 imputation 방법에 대해 알아보고, 그 중 mean imputation 방법을 사용하여 `자산총계`의 결측치를 처리해보세요.


```python
# 데이터를 새 변수에 저장 
df1_re = pd.read_excel(url, sheet_name= 0)
df2_re = pd.read_excel(url, sheet_name= 1)
```

```python
df1_re = df1_re.T
df2_re = df2_re.T

df1_re.columns = df_columns
df2_re.columns = df_columns

df1_re = df1_re[1:]
df2_re = df2_re[1:]
```

```python
# 자산총계 평균값 
df1['자산총계'].mean()
```

```python
df2['자산총계'].mean()
```

```python
# 자산총계 column의 결측치를 평균값으로 대체 
df1_re['자산총계'] = df1_re['자산총계'].fillna(df1['자산총계'].mean())
df1_re['자산총계']
```

```python
df2_re['자산총계'] = df2_re['자산총계'].fillna(df2['자산총계'].mean())
df2_re['자산총계']
```

##### **`부채 비율 2` 라는 새로운 피쳐를 생성하세요.**

> 부채 비율(%) = (부채총액 / 자본총액) x 100

위의 공식을 활용하여 `부채 비율 2` 라는 column을 생성합니다. 그리고, 생성한 column과 기존 `부채 비율` column의 값이 동일한지 확인해봅니다.

```python
df1['부채비율2'] = (df1['부채총계']/df1['자본총계'])*100
df1
```

```python
df2['부채비율2'] = (df2['부채총계']/df2['자본총계'])*100
df2
```

##### **`종목` 이라는 새로운 피쳐를 생성하세요.**

**`.repeat()`**를 활용하여 새로운 `종목`이라는 새로운 column을 생성합니다.
  - `df1`의 `종목` column에는 `008770` 문자열이 담기도록 합니다.  
  - `df2`의 `종목` column에는 `035250` 문자열이 담기도록 합니다. 

각각의 데이터 프레임에 `종목`이라는 column을 만든 후, 최종적으로 두 데이터프레임 `df1`과 `df2`를 **`.append()`**를 활용하여 합쳐주도록 합니다.

```python
import numpy as np

df1_new = np.repeat('008770', len(df1))
df2_new = np.repeat('035250', len(df2))
```

```python
df1['종목'] = df1_new
df2['종목'] = df2_new
```

```python
# 데이터프레임 합치기
df = df1.append(df2)
df
```
