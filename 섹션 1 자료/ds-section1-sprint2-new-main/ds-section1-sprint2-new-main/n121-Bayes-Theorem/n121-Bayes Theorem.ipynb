# **Part 1. 확률이란**

## **1-1. 전체 확률의 법칙**

확률이란, 어떠한 원인에 의해 사건이 발생할 때 이에 대한 결과를 예측하는 것을 말합니다.  
즉, 어떤 사건에서 어떤 일이 일어날 가능성을 **수**로 나타낸 것을 말합니다. 

확률을 이해하는 데 가장 좋은 예시는 **동전 던지기** 입니다. <br>
아래에서 동전 던지기 예를 통해 확률에 대해 배워보도록 합니다.

--------

동전을 던지는 행위는 **`앞면(H)`** 또는 **`뒷면(T)`**이라는 **`두 개의 잠재적 결과`**를 가지고 옵니다. <br>
하지만, 동전이 공평한 동전인지 구부러진 동전인지에 따라 앞면이 나올 확률은 달라집니다. 

  1. **공평한 동전** <br>
    - 만약 동전이 공평한 동전이라면 앞면이 나올 확률은 50% 이고, 뒷면이 나올 확률 또한 50% 입니다. <br>
       ``` 
       P(H) = 0.5
       P(T) = 0.5
       ```
  2. **구부러진 동전**
    - 만약 동전이 구부러진 동전이라면, 앞면이 나올 확률은 뒷면이 나올 확률보다 훨씬 더 커집니다.<br>
    - 만약 앞면이 나올 확률이 70%라면, 뒷면이 나올 확률은 무엇일까요?
      ```
      P(H) = 0.70
      P(T) = ?
      ```
    - 뒷면이 나올 확률은 30% 이어야 합니다. 

<br>

> **💡**  그 이유는 **전체 확률의 법칙**에 있습니다. <br>
전체 확률의 법칙이란, **`사건에 의해 발생한 모든 잠재적 결과들이 나올 확률의 합은 1이다`**입니다. 동전던지기는 앞면과 뒷면이라는 두 개의 결과를 초래하는 사건으로, 두 사건이 일어날 확률의 합은 1이어야 합니다. <br>
  ```
  P(A) = 1 - P(^A)
  ```







## **1-2. 확률 계산**

공평한 동전과 구부러진 동전 던지기를 통해 원하는 경우의 확률을 계산해 보도록 합니다. 

확률을 계산하는 방법에는 **두 가지**가 존재합니다. <br>
  - **`method_1.`** 모든 경우의 확률을 진리표를 통해 확인하고, 원하는 경우의 확률을 더해줍니다.
  - **`method_2.`** 확인하고자 하는 경우의 확률만을 곱하여 더해줍니다. 

우리는 아래 두 개의 시나리오 **a. 동전을 두 번 던졌을 때**와 **b. 동전을 세 번 던졌을 때**에 대하여 위의 두 가지 방법을 사용하여 확률을 확인하도록 합니다.

<br>

### **a. 동전을 두 번 던졌을 때의 확률** <br>
  - **공평한 동전** <br>
    - **`method_1.`** 공평한 동전을 두 번 던지는 행위의 **모든 잠재적 경우**와 **각 경우의 확률**은 아래와 같습니다. 앞면과 뒷면이 각각 한 번씩 나올 확률은 **2번**과 **3번**의 경우로, 이 두 경우의 확률을 더한 0.25 + 0. 25 = 0.5입니다.
  > <img width='500' src='https://i.imgur.com/VzhR8Ox.png'> 
  
    - **`method_2.`** 앞면이 나올 확률 **`P(H)은 0.5`**이고, 뒷면이 나올 확률 **`P(T)은 0.5`** 입니다. 앞면과 뒷면이 각각 한 번씩 나오는 경우의 수는 아래와 같이 두 가지이고, 두 경우의 0.25를 더한 0.5가 앞면과 뒷면이 각각 한 번씩 나올 확률입니다. <br>  
   
      **`P(H, T) = P(H) x P(T) = 0.5 x 0.5 = 0.25`** <br>
      **`P(T, H) = P(T) x P(H) = 0.5 x 0.5 = 0.25`** <br> 

<br>

  - **구부러진 동전**
    - **`method_1.`** 앞면이 나올 확률 **`P(H)는 0.7`**이고, **`뒷면이 나올 확률 P(T)는 0.3`**인 구부러진 동전을 두 번 던져 나올 수 있는 **모든 잠재적 경우**는 아래와 같고, 앞면과 뒷면이 각각 한 번씩 나올 확률은 **2번**과 **3번**의 경우로, 이 두 경우의 확률을 더한 0.21 + 0. 21 = 0.42 입니다.
    > <img width='500' src='https://i.imgur.com/oonIHnQ.png'>
    - **`method_2.`** 앞면과 뒷면이 각각 한 번씩 나오는 경우의 수는 아래와 같이 두 가지이고, 두 경우를 더한 0.42가 앞면과 뒷면이 각각 한 번씩 나올 확률입니다. <br>  
   
      **`P(H, T) = P(H) x P(T) = 0.7 x 0.3 = 0.21`** <br>
      **`P(T, H) = P(T) x P(H) = 0.3 x 0.7 = 0.21`** <br> 
<br>

> **💡 Tip** <br>
구부러진 동전은 앞면이 나올 확률이 높기 때문에, 앞면과 뒷면이 각각 한 번씩 나오는 확률은 공평한 동전을 던졌을 때보다 낮아졌음을 확인할 수 있습니다. 하지만 경우 1번의 두 번 다 앞면이 나올 확률은 공평한 동전을 던졌을 때인 0.25보다 0.49로 확률이 증가했음을 알 수 있습니다.        

<br>









### **b. 동전을 세 번 던졌을 때의 확률** <br>
  - **공평한 동전**
    - **`method_1.`** 공평한 동전을 세 번 던져 나올 수 있는 **모든 잠재적 경우**는 아래와 같습니다. 동전을 세 번 던져 앞면이 한 번만 나오는 경우는 **4번, 6번, 7번** 으로, 세 경우의 확률을 더한 0.125 + 0.125 + 0.125 = 0.375 가 답입니다.   
    > <img width='600' src='https://i.imgur.com/OOZLkOP.png'> <br>

    - **`method_2.`** 앞면이 한 번만 나오는 확률은 **`P(H, T, T)`**, **`P(T, H, T)`**, **`P(T, T, H)`** 로, 세 확률을 더한 0.375가 답입니다.  
    
      **`P(H, T, T) = P(H) x P(T) x P(T) = 0.5 x 0.5 x 0.5 = 0.125`** <br>
      **`P(T, H, T) = P(H) x P(T) x P(H) = 0.5 x 0.5 x 0.5 = 0.125`** <br>
      **`P(T, T, H) = P(H) x P(H) x P(T) = 0.5 x 0.5 x 0.5 = 0.125`** <br>
  
<br>

  - **구부러진 동전**
    - **`method_1.`** 구부러진 동전을 세 번 던져 나올 수 있는 **모든 잠재적 경우**는 아래와 같습니다. 동전을 세 번 던져 앞면이 한 번만 나오는 경우는 **4번, 6번, 7번** 으로, 세 경우의 확률을 더한 0.063 + 0.063 + 0.063 = 0.189 가 답입니다.  
    > <img width='600' src='https://i.imgur.com/n5n6rV6.png'> <br>
    - **`method_2.`** 세 번 던져 앞면이 나올 확률은 0.189 입니다.
      
      **`P(H, T, T) = P(H) x P(T) x P(T) = 0.7 x 0.3 x 0.3 = 0.063`**<br>
      **`P(T, H, T) = P(T) x P(H) x P(T) = 0.3 x 0.7 x 0.3 = 0.063`**<br>
      **`P(T, T, H) = P(T) x P(T) x P(H) = 0.3 x 0.3 x 0.7 = 0.063`**<br>


<br>

> ⚠️ <br>
위의 테이블에서 공평한 동전과 구부러진 동전 모두 **모든 경우의 확률을 더한 값은 1인 것을 확인**할 수 있습니다. <br> 
즉, 모든 결과의 확률을 더한 값은 1로써 이는 **전체 확률의 법칙**을 확인할 수 있는 부분입니다. 






# **Part 2. 이항 분포 (Binomial Distribution)**

````python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

probList = []

# 동전을 16번 던졌을 때 앞면이 나올 확률 
for i in range(17):
    res = math.factorial(16)/(math.factorial(i)*math.factorial(16-i))
    probList.append(res/2**16)

df = pd.DataFrame({'p':probList})

x = np.arange(0,17)
plt.bar(x, 'p', data=df)
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.title('Binomial Distribution of 16 Times Coin Flip')
plt.xticks(x);
````

### **2-1. 이항 분포란**

우리는 위에서 하나의 동전을 여러 번 던져 원하는 일련의 사건에 대한 확률을 구하는 방법에 대해 알아보았습니다. 이제 우리는 동전을 어려 번 던져 **원하는 사건의 모든 잠재적 경우**와 이에 대한 **확률**을 나타내는 **이항 분포**에 대해 배워보도록 합니다.

**이항분포**란 독립적으로 반복되어지는 행위에 의해서 **`실패`** 또는 **`성공`**과 같이 **`두 가지 옵션을 가지는 사건`**의 확률을 결정하는 함수입니다. 동전 던지기는 각 시행시 **`앞면`**과 **`뒷면`**이라는 **`두 가지 옵션`**을 가지는 사건으로 이항 분포의 대표적인 예입니다.

위의 그래프는 동전을 16번 던졌을 때 **앞면이 나오는 모든 잠재적 경우**와 이에 대한 **확률**을 나타내는 **이항 분포** 입니다. 동전을 던지는 독립적인 행위를 여러 번 시행해 앞면이 나오는 횟수가 **`X축`**이고, **`Y축`**은 앞면이 나오는 횟수에 대한 확률을 나타냅니다. 위의 그래프를 통해 동전을 16번 던졌을 때 앞면이 8번 나올 확률이 가장 높은 것을 알 수 있습니다.

<br>

**이항 분포**는 결과가 **`binary한 사건`**의 모든 잠재적 경우의 수와 이에 대한 확률을 보여줍니다.<br> 그러므로 데이터 분석에서 **소비자의 행동 패턴을 모델링** 하는 데 많이 이용됩니다. <br>
예를 들어, 소비자의 구매를 예측하는 모델링을 할 때, 소비자의 행동의 결과는 상품의 **`구매`** 또는 **`비구매`** 혹은 클릭을 눌러 **`전환`** 또는 **`비전환`** 등을 초래하는 두 가지인 경우가 많기에 이항 분포가 많이 사용됩니다. <br>

그럼, 이항 분포를 만들어 나가는 과정을 아래의 **동전 던지기 사건**을 통해 배워보도록 합시다!

#### **b-1. 사건의 경우의 수 확인**

  - **`method_2.`** 동전을 던졌을 때 원하는 사건이 차지할 수 있는 **`잠재적 위치의 개수`**를 세어 계산합니다. 동전을 던져 처음 앞면이 나오는 경우는 **`다섯 개의 잠재적 위치`**에 존재할 수 있습니다. 그리고 두 번째 앞면이 나오는 경우는 첫 번째 앞면에 의해 차지된 하나의 위치를 제외한 **`네 개의 잠재적 위치`**에 존재할 수 있습니다. 이를 곱한 뒤, 중복될 수 있는 경우의 수를 나누어 줍니다.

```
Product of Available Positions/Product of Duplication = n!/k!(n-k)!
```

##**2-3. 이항 분포 Formula**

위에서 우리는 동전 던지기와 같은 각 사건이 독립적이고, 두 가지의 결과를 가지는 사건의 확률을 결정하는 이항 분포의 다양한 경우와 그 확률을 확인해 보았습니다. 

그러면 **이항분포의 확률**을 도출하는 데 **일관적인 공식**을 아래에서 확인하도록 합니다.

<br>

> **💡** 동전을 **`n번`** 던져 앞면이 **`k번`** 나올 확률: <br>
>
>**`1.`** n번 동전을 던저 앞면이 k번 나오는 경우의 수: $n! / k!(n-k)!$ <br>
**`2.`** k번 앞면이 나오고 n-k번 뒷면이 나오는 확률: ${p^k}{(1-p)^{(n-k)}}$ <br>
**`3.`** 이항 분포의 Formula는 **`1.`**의 **n번 동전을 던저 앞면이 k번 나오는 경우의 수** 에 **`2.`**의 **k번 앞면이 나오고 n-k번 뒷면이 나오는 확률**을 곱한 아래의 형태입니다. <br>
$$n! / k!(n-k)! \times {p^k}{(1-p)^{(n-k)}}$$


# **Part 3. 조건부 확률 (Conditional Distribution)**

## **3-1. 조건부 확률이란**
일상적인 상황에서 사건들은 서로 연관되어 있습니다. **조건부 확률**이란 어떠한 사건의 결과에 의해 영향을 받는 한 사건의 결과에 대한 확률을 일컫습니다. 

<img width='300' src='https://i.imgur.com/teH9ho6.png'>

위의 이미지가 나타내는 것은 보자기 안에 **`공평한 동전`**과 **`구부러진 동전`** 두 개가 존재하고,<br>
우리는 이 보자기에서 동전을 꺼내 던졌을 때 앞면이 나오는 확률을 확인하려고 합니다.  
이때 동전을 던져 앞면이 나올 확률이 무엇인지 확인하기 전에, 보자기에서 동전을 꺼내는 사건을 살펴보아야 합니다. 그러므로 이 시나리오는 **`조건부 확률의 예`**입니다. <br>
그 이유는, 동전을 던져 앞면이 나올 확률은 내가 보자기에서 어떤 동전을 선택하느냐에 따라 달라지기 때문입니다.   

아래의 두 case를 통해서 조건부 확률에 대해 배워보도록 합니다. <br>

  - **`CASE 1.`** 보자기에서 동전을 꺼내, 동전을 한 번 던져 앞면이 나올 확률 <br>
  - **`CASE 2.`** 보자기에서 동전을 꺼내, 동전을 두 번 던져 앞면과 뒷면이 나올 확률 <br>

<br>

### **CASE 1**


하나의 보자기에 **`공평한 동전`**과 **`구부러진 동전`** 두 개의 동전이 있습니다. <br>
공평한 동전을 던져 앞면이 나올 확률은 **`0.5`** 이고, 구부러진 동전을 던져 앞면이 나올 확률은 **`0.7`** 입니다.

우선, 이 가방에서 동전 하나를 꺼냅니다. 공평한 동전과 구부러진 동전이 pick 될 확률은 각각 **`0.5`**로 같습니다. <br>
그럼 동전 하나를 꺼내 동전을 던졌을 때, 이 동전의 앞면이 나올 확률은 몇일까요?

이 경우 동전을 던져 앞면이 나올 확률은, 동전 던지기 사건 전인 보자기에서 어떠한 동전을 꺼냈는지에 따라 확률이 달라지므로 우리는 조건부 확률을 통해 구해보도록 합니다.

<img width='700' src='https://i.imgur.com/eE7VR88.png'>

위의 테이블에서 동전 앞면이 나오는 경우는 **1**과 **3**으로 두 확률을 합치면 **`0.6`**입니다. <br>
이 확률은 두 종류의 동전이 보자기에 있을 때 앞면을 꺼낼 확률로서, 두 개의 공평한 동전이 있을 때 동전을 꺼내 앞면이 나올 확률 **`0.5`** 에 비하면 앞면을 꺼낼 확률이 더 높아진 것을 알 수 있습니다. 

위의 시나리오를 공식으로 표현해 보겠습니다. <br>
**P(앞면|공평한동전)P(공평한동전) + P(앞면|구부러진동전)P(구부러진동전)** <br>
**`= 0.5 x 0.5 + 0.7 x 0.5 = 0.6`**

<br>

### **CASE 2**

이번 case에서 우리는 보자기에서 동전 하나를 꺼내, 동전을 두 번 던집니다. <br> 
첫 번째 던지기에서는 **`앞면`**이 나오고, 두 번째 던지기에서는 **`뒷면`**이 나올 확률은 몇인가요? 

<img width='700' src='https://i.imgur.com/EsrOAhy.png'>
  
위에서 경우 **2**와 **6**이 해당됩니다. 따라서 두 확률을 더한 0.125 + 0.105 = 0.23 입니다. <br>

위의 시나리오를 공식으로 표현하면 아래와 같습니다. <br>
**P(공평한동전)P(앞면|공평한동전)P(뒷면|공평한동전) + P(구부러진동전)P(앞면|구부러진동전)P(뒷면|구부러진동전)** <br>
**`= 0.5 x 0.5 x 0.5 + 0.5 x 0.7 x 0.3 = 0.125 + 0.105 = 0.23`**

























## **3-2. 조건부 확률의 예**

위에서 우리는 동전을 보자기에서 꺼내 던져 나오는 확률을 통해 조건부 확률을 배웠습니다. <br>
이제 우리는 암진단 테스트를 이용해 조건부 확률을 배워보도록 합니다. <br>

암진단 테스트의 결과가 **`positive`**로 또는 **`negative`**로 나오는 결과는, **환자가 암이 있는지 없는지에 따라 달라집니다**. 그러기에 이도 대표적인 **조건부 확률의 예**입니다.  

<br> 

> **🥇 시나리오** <br> 
전체 인구중 10% 정도가 걸리는 암이 있습니다. 암이 있는 사람중 90% 는 테스트에서 **`positive`**하게 나옵니다. 그리고 암이 없는 사람 중 80%는 테스트에서 **`negative`**하게 나옵니다.
- **`P(C)`** = 0.1
- **`P(Pos|C)`** = 0.9
- **`P(Neg|^C)`** = 0.8

<br>

- **`P(C)`**란 암에 걸릴 확률을 말합니다.
- **`P(Pos|C)`**란 암에 걸린 환자 중 테스트를 통해 암에 걸렸다고 나올 확률을 말합니다.
- **`P(Neg|^C)`**란 암에 걸리지 않은 환자 중 테스트를 통헤 암에 걸리지 않았다고 나올 확률을 말합니다. <br>

<br>

> **💡 Tip** <br> 위의 시나리오에서 **전체 확률의 법칙**에 의해 우리는 아래의 사항 또한 유추할 수 있습니다.

- **`P(^C)`** = 0.9
- **`P(Neg|C)`** = 0.1
- **`P(Pos|^C)`** = 0.2



<br>

> <img width='300' src='https://i.imgur.com/tP2QWiD.png'>

<br>

위의 시나리오의 데이터를 통해 아래의 질문에 대답해 보도록 합니다.

  - **`Q1.`** 암이 있고, 암진단 테스트에서 Positive로 나올 확률은 무엇인가요? <br>
  P(C $\cap$ Pos) = P(Pos|C) x P(C) = 0.1 x 0.9 = 0.09<br>

  - **`Q2.`** 암이 있고, 암진단 테스트에서 Negative로 나올 확률은 무엇인가요? <br>
    P(C $\cap$ Neg) = P(Neg|C) x P(C) = 0.1 x 0.1 = 0.01 <br>

  - **`Q3.`** 암이 없고, 암진단 테스트에서 Positive로 나올 확률은 무엇인가요? <br>
    P(^C $\cap$ Pos) = P(Pos|^C) x P(^C) 0.9 x 0.2 = 0.18 <br>

  - **`Q4.`** 암이 없고, 암진단 테스트에서 Negative로 나올 확률은 무엇인가요? <br>
    P(^C $\cap$ Neg) = P(Neg|^C) x P(^C) = 0.8 x 0.9 = 0.72 <br>
  
  - **`Q5.`** 암이 있던 없던, 전체 모집단에서 암진단에서 암이 있다고 나올 확률은 무엇인가요? 

    P(Pos) = P(C) x P(Pos|C) + P(^C) x P(Pos|^C) = 0.1 x 0.9 + 0.9 x 0.2 = 0.27

<br>



## **3-3. 조건부 확률 Formula**

조건부 확률에서의 요점은, <br>
한 사람에게 질병이 있는지 확인하는 테스트의 결과는, 이 사람에게 **정말로 질병이 있는지 없는지에 따라 달라지는 사건이다** 라는 것입니다. 그러므로 조건부 확률은 동전 던지기와 같이 독립적인 시행이 아닌, **서로 관련되어 있는 사건**에 사용됩니다. <br>

예를 들어, 의료 테스트에서의 결과는 테스트 대상자의 질병의 유무에 의해 달라지고, 이는 우리가 확률적 추론을 할 때 고려해야 할 사항입니다. <br> 

그 이유는, 테스트에서 환자가 Positive한 결과를 받는 확률은, 그 사람이 정말로 질병이 있는지 없는지에 따라 달라지기 때문입니다. <br> 

즉, 만약 환자에게 질병이 있다면 테스트 결과가 Positive가 나올 가능성이 높아집니다. 

<br>

> **💡 Tip** <br> 
위에서 찾은 조건부 확률 예의 위치를 바꿔주며 아래의 결과를 얻을 수 있습니다. <br>
- $P(Pos|C) = P(C \cap  Pos) / P(C)$
- **|** 는 **`given`** 의 의미이고 **$\cap$** 는 **`and`**의 의미 입니다. 
- 즉, 어떠한 사람이 질병이 있다는 전제 하에 테스트에서 positive로 나올 확률은, 어떠한 사람이 **질병이 있고 테스트에서 postive로 나올 확률**을 **질병을 가질 확률**로 나눈 것과 같습니다.

<br>

이를 아래와 같이 A와 B라는 두 사건의 조건부 확률로 공식화 할 수 있습니다. <br>
- $P(A|B) = P(A \cap  B) / P(B)$







# **Part 4. Bayesian Theorem**

위에서 배운 조건부 확률을 베이지안 이론으로 확장시켜 보도록 합니다.

Bayesian Theorem의 공식은 아래와 같습니다.

$P(A|B) = P(A \cap  B) / P(B) = P(B|A)P(A) / P(B)$

<br>

Bayesian Theorem을 위에서 배운 암을 진단하는 테스트에 적용하여 이해해보도록 합니다.

- **`P(Pos|C)`** = 0.9
- **`P(Neg|^C)`** = 0.9

$P(C|Pos) = P(C \cap  Pos) / [P(Pos ∩ C) + P(Pos ∩ no C)]$

- Posterior Probability  = Prior Probability (테스트를 진행하기 전의 확률) x Evidence (테스트로부터의 증거)
